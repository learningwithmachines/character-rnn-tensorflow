```python
#### Exercise: Recursive Sequence Extraction
from matplotlib import pyplot as plt
import numpy as np
from functools import lru_cache
%matplotlib inline
# define generator

@lru_cache(maxsize=8192)
def memoize(f):
    memo = {}
    def helper(*args):
        if args not in memo:
            memo[args] = f(*args)
        return memo[args]
    return helper

@memoize
def genseq(seed,omega, idx):
    # first order generator
    recur  = lambda : genseq(seed,omega,idx-1)
    if idx == 1:
        return seed
    else:
        return omega*recur() - omega*recur()*recur()
```

```python
seq_x = np.linspace(1,50,50)
seq_y = [genseq(0.5,1,x) for x in seq_x]

plt.plot(seq_x, seq_y, marker='d', markerfacecolor='r', color='b')
```
